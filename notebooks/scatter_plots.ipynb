```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import combinations

# List of model names
model_names = ["Model eos2ta5", "Model eos30f3", "Model eos30gr", "Model eos43at", "Model eos4tcc"]

# List of CSV file paths for each model
csv_files = ["../data/eos2ta5/prediction_eos2ta5.csv", "../data/eos30f3/prediction_eos30f3.csv", "../data/eos30gr/prediction_eos30gr.csv", "../data/eos43at/prediction_eos43at.csv", "../data/eos4tcc/prediction_eos4tcc.csv"]

# Create an empty list to store DataFrames for each model
dfs = []

# Load each CSV file into a DataFrame
for file_path in csv_files:
    df = pd.read_csv(file_path)
    dfs.append(df)

# Create combinations of models for pairwise comparison
model_combinations = list(combinations(range(len(dfs)), 2))

# Plotting scatter plots for pairwise comparisons
fig, axes = plt.subplots(len(model_combinations)//2, 2, figsize=(12, 8))
axes = axes.flatten()

for idx, (model1, model2) in enumerate(model_combinations):
    ax = axes[idx]
    ax.scatter(dfs[model1]['probability'], dfs[model2]['probability'])
    ax.set_title(f'{model_names[model1]} vs {model_names[model2]}')
    ax.set_xlabel(f'{model_names[model1]} Probability')
    ax.set_ylabel(f'{model_names[model2]} Probability')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
